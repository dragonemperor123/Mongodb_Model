```python
from bson.binary import Binary
from bson import ObjectId
import pymongo, gridfs
from gridfs import GridFS
from pymongo import MongoClient

MONGO_HOST = "127.0.0.1"
MONGO_PORT = 27017
MONGO_DB = "FirstDB"

model_file = 'model.pkl'

myclient = pymongo.MongoClient("mongodb+srv://Adwait:Darkmatter!1@cluster0.g1q8onc.mongodb.net/?retryWrites=true&w=majority")
mydb = myclient[MONGO_DB]
mycol = mydb['Addy']

# Store model as binary document
with open(model_file, "rb") as f:
    encoded = Binary(f.read())
mycol.insert_one({"filename": model_file, "file": encoded, "description": "Keras model" })

# Retrieve and store the ML model from MongoDB
data = mycol.find_one({'filename': 'model.pkl'})
with open("keras_model_fromMongo.h5", "wb") as f:
    f.write(data['file'])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report


logreg = LogisticRegression()

# Fit the classifier to the training data
logreg.fit(X_train, Y_train)

# Predict the labels of the test set: y_pred
y_pred = logreg.predict(X_test)

# Compute and print the confusion matrix and classification report
print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```
